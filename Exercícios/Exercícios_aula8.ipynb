## Exercicio 1:

Resolução numérica de equações diferenciais. Imagine que tem um/a maratonista treinando, que num trecho de 5km tem velocidade quase constante de 12km/h. Produza uma lista com a distância percorrida em intervalos de 1 minuto.

```python
velocidade = 12 * (1000/60) # velocidade em m/min
distância = 5 * 1000 # distância em metros
tempo1 = distância/velocidade
print("Esse percusso demora:", tempo1, "minutos")

list = [0]
def distância_percorrida(distância, velocidade):
    for _ in range(int(tempo1)):
        nova_distância = list[-1] + velocidade
        list.append(nova_distância)
        
distância_percorrida(distância, velocidade) 
print("Os valores da distancia percorrida em intervalos de 1 minuto é: ", list)
```

## Exercicio 2:

O/A mesmo/a atleta no seguinte trecho acelera nos seguintes 200 metros até chegar numa velocidade de 15km/h, e mantém esta velocidade por 1,8km. Queremos obter uma tabela ou gráfico da distância percorrida en função do tempo, e calcular e o tempo total para chegar em 7km, faça isto utilizando listas.

```python
velocidade_i = 12 * (1000/60)  # velocidade inicial em m/min
distancia_2 = 200  # metros
velocidade_f = 15 * (1000/60)  # velocidade final em m/min

# Por Torricelli:
aceleracao = (velocidade_f**2 - velocidade_i**2) / (2 * distancia_2)

# Achando o tempo supondo aceleração constante
tempo2 = (velocidade_f - velocidade_i) / aceleracao
print("Tempo necessário nesse percurso de 200 metros é:", tempo2, "minutos")

list = [0]
def distância_percorrida(distância, velocidade_i, aceleracao):
    for _ in range(round(tempo2)):
        nova_distância = list[-1] + velocidade_i + (1/2)*aceleracao
        list.append(nova_distância)
        
distância_percorrida(distância, velocidade_i, aceleracao) 
print("Os valores da distancia percorrida em intervalos de 1 minuto são: ", list)

#percuso final:

velocidade_final = 15 * (1000/60)  # velocidade em m/min
distancia_final = 1.8 * 1000  # distância em metros
tempo3 = distancia_final / velocidade_final
print("Esse percurso final demora:", tempo3, "minutos")

lista_final = [0]

def distancia_percorrida_final(distancia_final, velocidade_final):
    for _ in range(int(tempo3)):
        nova_distancia = lista_final[-1] + velocidade_final
        lista_final.append(nova_distancia)

distancia_percorrida_final(distancia_final, velocidade_final)
print("Os valores da distância percorrida em intervalos de 1 minuto são: ", lista_final)
```
